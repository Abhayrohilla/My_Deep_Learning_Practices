```python
import re 
def remove_urls(text):
    url_pattern = re.compile(r'https?://\S+|www\.\S+')
    return ''.join([url_pattern.sub(r'', text)])
Remove_urls = X.apply(lambda text: remove_urls(text))
```

```python
Remove_urls
```

```python
import string

PUNCT_TO_REMOVE= list(string.punctuation)
def remove_punct(text):
    return ''.join([i for i in text if i not in PUNCT_TO_REMOVE])
Remove_punct = Remove_urls.apply(lambda text: remove_punct(text))
```

```python
Remove_punct
```

```python
from nltk.corpus import stopwords
STOPWORDS = set(stopwords.words('english'))
def romve_stopword(text):
    return ' '.join([i for i in text.split() if i not in STOPWORDS])
Remove_stopword = Remove_punct.apply(lambda text: romve_stopword(text))
```

```python
Remove_stopword
```

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
def lemmatize_words(text):
    return " ".join([lemmatizer.lemmatize(word) for word in text.split()])

lemmatize_word = Remove_stopword.apply(lambda text: lemmatize_words(text))
```

```python
lemmatize_word
```

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()
def Text_stemmer(text):
    return ' '.join([ps.stem(w) for w in text.split()])
Text_Stemmer = Remove_stopword.apply(lambda text: Text_stemmer(text))
```
